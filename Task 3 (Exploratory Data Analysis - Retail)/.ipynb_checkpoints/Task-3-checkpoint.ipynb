## Data Analysis

```python
sns.pairplot(df)
```

```python
fig,axes = plt.subplots(1,1,figsize=(9,6))
sns.heatmap(df.corr(), annot= True)
plt.show()
```

### from above heatmap we can see thar 'Sales' and 'Profit(our target)' have best corelation(0.48) of all other attributes, then comes the Quantity, and so on 

```python
df['Ship Mode'].value_counts()
```

```python
sns.countplot(x=df['Ship Mode'])
```

```python
df['Segment'].value_counts()
```

```python
sns.countplot(x=df['Segment'])
```

```python
sns.catplot("Ship Mode", hue="Segment", data=df, kind="count", aspect=1.5, palette="Set1")
```

### most of shipment is sent through 'Standard Class' and to the 'Consumers' most of the time. So, standard shipping to consumer is realy important aspect. 

```python
print(df['State'].value_counts())
plt.figure(figsize=(15,8))
sns.countplot(x=df['State'])
plt.xticks(rotation=90)
plt.show()
```

### most of the products are sent to 'California', 'New York' and 'Texas' in US.

```python
df.groupby("State").Profit.agg(["sum","mean","min","max","count","median","std","var"])
```

```python
plt.figure(figsize=[24,15])
ax = sns.barplot(x="State", y="Profit", data=df, palette="Set1",)
plt.xticks(rotation=90, fontsize=16)
plt.yticks(fontsize=15)
plt.title("States VS Profit",fontsize=24)
plt.xlabel("States",fontsize=20)
plt.ylabel("Profit",fontsize=20)
plt.tight_layout()
```

### the best profit levels are coming from Vermount, Montana and Indiana, and at the same time products of  Colorado, ohio and North California are in Loss.

```python
sns.countplot(x=df['Region'])
```

```python
plt.figure(figsize=(16,8))
plt.bar('Sub-Category','Category', data=df)
plt.show()
```

```python
plt.figure(figsize=[12,8])
ax = sns.barplot(x="Category", y="Profit", data=df, palette="Set1")
```

```python
plt.figure(figsize=(40,25))
sns.barplot(x=df['Sub-Category'], y=df['Profit'])
```

```python
fig, ax = plt.subplots(figsize = (10 , 6))
ax.scatter(df["Sales"] , df["Profit"])
ax.set_xlabel('Sales')
ax.set_ylabel('Profit')
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (10 , 6))
ax.scatter(df["Discount"] , df["Profit"])
ax.set_xlabel('Discount')
ax.set_ylabel('Profit')
plt.show()
```

```python
plt.figure(figsize=[12,8])
ax = sns.barplot(x="Sub-Category", y="Profit",hue="Region", data=df, palette="viridis")
plt.xlabel("Subcategories",fontsize=15)
plt.ylabel("Profit",fontsize=15)
```
